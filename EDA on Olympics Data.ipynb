```python
#taking those indexes to replace region with notes
x=country[pd.isna(country.region)].index
#changing the na's
for i in x:
    country.region[i]=country.notes[i]
```

```python
#now see country data
country
```

```python
#merging both datasets
data=data.merge(country,how='outer',on='NOC')
#remove unnecessary column
data.drop('notes',axis=1,inplace=True)
#renaming region column
data=data.rename(columns={'region':'Country'})
#now take a look on dataset
data
```
